```python
data_full_path_hcpya = root + '/images/' + datasets[0] + '/' + model_type +'/'+ model_details + '/' + model_test_type  # Replace with your target directory
hcpya_performance = np.load(f'{data_full_path_hcpya}/list_of_models_predtarget_rho.npy', allow_pickle=True)

data_full_path_abcd = root + '/images/' + datasets[1] + '/' + model_type +'/'+ model_details + '/' + model_test_type # Replace with your target directory
abcd_performance = np.load(f'{data_full_path_abcd}/list_of_models_predtarget_rho.npy', allow_pickle=True)


fig, axes = plt.subplots(figsize=(18, 12))
# axes = axes.flatten()

# Flatten values
tr_demean_vals1 = hcpya_performance[0].flatten()
tr_orig_vals1   = hcpya_performance[1].flatten()
te_demean_vals1 = hcpya_performance[2].flatten()
te_orig_vals1   = hcpya_performance[3].flatten()

tr_demean_vals2 = abcd_performance[0].flatten()
tr_orig_vals2   = abcd_performance[1].flatten()
te_demean_vals2 = abcd_performance[2].flatten()
te_orig_vals2   = abcd_performance[3].flatten()

# Concatenate all RHO values
rho_vals = np.concatenate([
    tr_demean_vals1, tr_orig_vals1, te_demean_vals1, te_orig_vals1,
    tr_demean_vals2, tr_orig_vals2, te_demean_vals2, te_orig_vals2
])

condition_labels = (
    ["tr_dmean"] * len(tr_demean_vals1) +
    ["tr_org"]   * len(tr_orig_vals1) +
    ["te_dmean"] * len(te_demean_vals1) +
    ["te_org"]   * len(te_orig_vals1) +
    ["tr_dmean"] * len(tr_demean_vals2) +
    ["tr_org"]   * len(tr_orig_vals2) +
    ["te_dmean"]* len(te_demean_vals2) +
    ["te_org"]   * len(te_orig_vals2)
)

dataset_labels = (
    ["HCPYA"] * (len(tr_demean_vals1) + len(tr_orig_vals1) + len(te_demean_vals1) + len(te_orig_vals1)) +
    ["ABCD"] * (len(tr_demean_vals2) + len(tr_orig_vals2) + len(te_demean_vals2) + len(te_orig_vals2))
)

# Construct long-form DataFrame
df = pd.DataFrame({
    "RHO values": rho_vals,
    "Performance": condition_labels,
    "Dataset": dataset_labels
})

sns.violinplot(data=df, x="Performance", y="RHO values", hue="Dataset", inner="point", ax=axes, palette="pastel")

axes.set_title("Violin Plots Across Datasets")
axes.set_ylabel("RHO Values")
axes.axhline(0, color='black', linestyle='--', linewidth=1)
axes.legend(title="Dataset")

# Save figure
filename = '/model_rho_performance_grouped.png'
# print(f'Saving to path:{directory}')
# plt.savefig(directory + filename)
plt.show()
plt.close()
```
